```python
"""
Un dispositivo de cilindro-pistón contiene 0.41 kg de **aire**, en un principio a 1.5 MPa y 430 °C. Primero se 
expande el aire isotérmicamente hasta 600 kPa, y después se comprime en un proceso politrópico con un 
exponente politrópico de 1.5, hasta la presión inicial; por último, se comprime a presión constante hasta 
llegar al estado inicial. Determine el trabajo de frontera para cada proceso, y el trabajo neto del ciclo
"""
import math as mt
import numpy as np
import matplotlib.pyplot as plt


def Proceso1 (P0,V0,P1) -> float :
  return P0*V0/P1
def Proceso2 (P1,V1,P2)-> float:
  v2 = pow(  P1*pow(V1,1.5)/P2 ,1/1.5)
  return v2

#Datos
masa = 0.41 # Kg
Presion0 = 1.5e3 # KPa
Temp0 = 430 + 273 # Kelvin

Presion1 = 600 # KPa
Presion2,Presion3 = [Presion0]*2  # KPa

rAire = 0.287  # KJ /Kg*K

volumen0 = (masa*rAire*Temp0)/(Presion0) #m^3
print(f"El volumen Inicial es: {round(volumen0,3)}  m^3")

volumen1 = Proceso1(Presion0,volumen0,Presion1)  #m^3
print(f" EL volumen en el estado 1 es {round(volumen1,3)} m^3")


volumen2 = Proceso2(Presion1,volumen1,Presion2) #$M^3$
print(f"El volumen del estado 2 es {volumen2} m^3")

presionesProceso1 = [p for p in np.arange(Presion0,Presion1-1,-50)]  # En KPa
volumenesProceso1 = [Proceso1(Presion0,volumen0,p) for p in presionesProceso1] #En m^3

presionesProceso2 = [p for p in np.arange(Presion1,Presion2+1,50)]
volumenesProceso2 = [Proceso2(Presion1,volumen1,p) for p in presionesProceso2] #m^3

volumenesProceso3 = [v for v in np.arange(volumen2,volumen0-0.001,-0.001)]
presionesProceso3 = [ Presion0 for v in volumenesProceso3]

plt.title("Diagrama P-V")
plt.ylabel("Presion KPa")
plt.xlabel("Volumen [$m^3$]")
plt.plot(volumenesProceso1,presionesProceso1,Linewidth=2.5, linestyle="-", label="Proceso 1 ")
plt.plot(volumenesProceso2,presionesProceso2,Linewidth=2.5, linestyle="-", label="Proceso 2")
plt.plot(volumenesProceso3,presionesProceso3,label = 'Proceso 3')
plt.annotate("Estado 1",xy=(volumen1,Presion1),xytext=(volumen1-0.009,Presion1+100),xycoords="data",arrowprops = dict(arrowstyle="->",connectionstyle="arc3,rad=0.2"))
plt.annotate("Estado 0",xy=(volumen0,Presion0),xytext=(volumen0+0.009,Presion0-100),xycoords="data",arrowprops = dict(arrowstyle="->",connectionstyle="arc3,rad=0.2"))
plt.annotate("Estado 2",xy=(volumen2,Presion2),xytext=(volumen2+0.009,Presion2-100),xycoords="data",arrowprops = dict(arrowstyle="->",connectionstyle="arc3,rad=0.2"))
plt.grid()
plt.legend(loc='upper right')
plt.show()
```
